```python
import random

class Player:
    def __init__(self, name):
        self.name = name
        self.health = 100
        self.inventory = []

    def take_damage(self, damage):
        self.health -= damage

    def is_alive(self):
        return self.health > 0

    def add_to_inventory(self, item):
        self.inventory.append(item)

class Game:
    def __init__(self):
        self.player = None

    def start(self):
        print("Welcome to the Adventure treasure hunt in a cave!")
        player_name = input("Enter your name: ")
        self.player = Player(player_name)
        self.intro()

    def intro(self):
        print(f"Hello, {self.player.name}! Your Hunt begins now.")
        print("You find yourself at a crossroad. You can go left or right.")

        while self.player.is_alive():
            choice = input("Will you go left or right? (left/right): ").lower()
            if choice == "left":
                self.left_path()
            elif choice == "right":
                self.right_path()
            else:
                print("Invalid choice. Please enter 'left' or 'right'.")

        print("Game over! Your adventure has come to an end.")

    def left_path(self):
        print("You chose to go left. You encounter a friendly merchant.")
        print("The merchant offers you a healing potion. Do you want to buy it?")
        choice = input("Type 'buy' or 'ignore': ").lower()

        if choice == "buy":
            self.player.add_to_inventory("Healing Potion")
            print("You now have a Healing Potion in your inventory.")
        elif choice == "ignore":
            print("You continue your journey without buying the potion.")
        else:
            print("Invalid choice. The merchant leaves.")

    def right_path(self):
        print("You chose to go right. You stumble upon a dark cave.")
        print("Inside the cave, you find a treasure chest guarded by a dragon!")
        print("You must fight the dragon to get the treasure.")

        while self.player.is_alive():
            fight_choice = input("Do you want to 'fight' or 'run' from the dragon? ").lower()

            if fight_choice == "fight":
                dragon_health = random.randint(50, 100)
                while dragon_health > 0:
                    player_attack = random.randint(10, 30)
                    dragon_health -= player_attack
                    print(f"You dealt {player_attack} damage to the dragon. Dragon's health: {dragon_health}")
                    if dragon_health <= 0:
                        print("You defeated the dragon and obtained the treasure chest!")
                        self.player.add_to_inventory("Treasure Chest")
                        break

                break
            elif fight_choice == "run":
                print("You ran away from the dragon, back to the crossroad.")
                break
            else:
                print("Invalid choice. Please enter 'fight' or 'run'.")

if __name__ == "__main__":
    game = Game()
    game.start()
```
